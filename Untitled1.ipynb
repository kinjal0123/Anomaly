```python
import pandas as pd
import numpy as np
import requests

np.random.seed(42)
```

```python
keywords = ["Aloe vera juice", "Fruit drink", "Curcumin C3", "Graviola", "Ecoensulin"]

rows_per_keyword = 400  # 5 keywords × 400 = 2000 rows
dates = pd.date_range(end=pd.Timestamp.today(), periods=rows_per_keyword)
```

```python
rows = []

for keyword in keywords:
    base_impressions = np.random.randint(1000, 5000)
    inventory = np.random.randint(50,150)

    for i in range(rows_per_keyword):
        trend = 1 + (i * 0.001)
        impressions = int(base_impressions * trend * np.random.uniform(0.9,1.1))
        ctr = np.random.uniform(0.02, 0.08)
        clicks = int(impressions * ctr)
        cpc = np.random.uniform(5,15)
        spend = clicks * cpc
        cvr = np.random.uniform(0.1,0.3)
        orders = int(clicks * cvr)
        avg_order_value = np.random.uniform(300,800)
        sales = orders * avg_order_value
        bid = cpc * np.random.uniform(0.9,1.2)
        acos = spend / sales if sales>0 else 0
        roas = sales / spend if spend>0 else 0
        inventory -= orders
        stock_out_flag = 1 if inventory <= 0 else 0
        inventory = max(inventory,0)
        organic_rank = np.random.randint(1,100)

        rows.append([
            dates[i], keyword, impressions, clicks, spend, sales,
            orders, ctr, cpc, cvr, acos, roas, bid,
            inventory, stock_out_flag, organic_rank
        ])
```

```python
df = pd.DataFrame(rows, columns=[
    "date","keyword","impressions","clicks","spend","sales",
    "orders","ctr","cpc","cvr","acos","roas","bid",
    "inventory_level","stock_out_flag","organic_rank"
])
```

```python
df = df.sort_values(by=["keyword","date"])

for col in ["impressions","clicks","spend","sales","orders","ctr","cvr","acos","cpc"]:
    df[f"{col}_change_1d"] = df.groupby("keyword")[col].pct_change()

df = df.fillna(0)
```

```python
def fetch_rainforest_data(keyword):
    url = "https://api.rainforestapi.com/request"
    params = {
        "api_key": API_KEY,
        "type": "search",
        "amazon_domain": "amazon.in",
        "search_term": keyword
    }
    try:
        response = requests.get(url, params=params)
        data = response.json()
        
        # Rainforest API ka actual path 'search_results' list hota hai
        all_results = data.get("search_results", [])
        
        sponsored_count = 0
        organic_count = 0
        prices = []
        ratings = []

        for item in all_results:
            # Check if item is sponsored
            if item.get("sponsored", False):
                sponsored_count += 1
            else:
                organic_count += 1
                # Organic products ka price aur rating lena
                if "price" in item and "value" in item["price"]:
                    prices.append(item["price"]["value"])
                if "rating" in item:
                    ratings.append(item["rating"])

        avg_price = np.mean(prices) if prices else 0
        avg_rating = np.mean(ratings) if ratings else 0
        
        return {
            "competitor_avg_price": avg_price,
            "sponsored_count": sponsored_count,
            "organic_count": organic_count,
            "competitor_avg_rating": avg_rating
        }
    except Exception as e:
        print(f"Error fetching {keyword}: {e}")
        return {
            "competitor_avg_price": 0, "sponsored_count": 0,
            "organic_count": 0, "competitor_avg_rating": 0
        }
```

```python
market_data = {}

for kw in keywords:
    print(f"Fetching competitor data for {kw}...")
    market_data[kw] = fetch_rainforest_data(kw)
```

```python
market_df = pd.DataFrame.from_dict(market_data, orient='index').reset_index()
market_df.rename(columns={"index":"keyword"}, inplace=True)

market_df.head()
```

```python
combined_df = df.merge(market_df, on="keyword", how="left")
```

```python
combined_df["competition_intensity"] = combined_df["sponsored_count"] / (
    combined_df["sponsored_count"] + combined_df["organic_count"]
)

combined_df["price_gap"] = combined_df["bid"] - combined_df["competitor_avg_price"]

combined_df["market_pressure_score"] = combined_df["cpc_change_1d"].fillna(0) + combined_df["competition_intensity"]

combined_df.fillna(0, inplace=True)
```

```python
combined_df.to_csv("final_2000_combined_dataset.csv", index=False)

print("Final combined dataset (~2000 rows) ready for anomaly detection!")
```

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

def detect_business_anomalies(df):
    
    df.columns = df.columns.str.strip()
    
    
    desired_features = [
        'acos', 'cvr', 'cpc', 'sales_change_1d', 
        'price_gap', 'market_pressure_score', 'organic_rank', 'inventory_level'
    ]
    
    
    available_features = [f for f in desired_features if f in df.columns]
    missing_features = [f for f in desired_features if f not in df.columns]
    
    if missing_features:
        print(f"⚠️ Warning: Missing columns {missing_features}. Model will use: {available_features}")
    
    if not available_features:
        raise ValueError("not found anu important feature!")

    
    df_model = df[available_features].fillna(0)
    
    # Scaling
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(df_model)
    
    # Isolation Forest
    model = IsolationForest(contamination=0.05, random_state=42)
    df['anomaly_signal'] = model.fit_predict(X_scaled)
    
    return df[df['anomaly_signal'] == -1].copy()


try:
    detected_anomalies = detect_business_anomalies(combined_df)
    print(f"Success! Found {len(detected_anomalies)} anomalies.")
except Exception as e:
    print(f"error : {e}")
```

```python
def explain_anomaly(row):
    reasons = []
    # Negative Anomalies (Bad News)
    if row['sales_change_1d'] < -0.3:
        reasons.append(f"Sales dropped significantly ({row['sales_change_1d']:.1%})")
    if row['inventory_level'] < 10:
        reasons.append("Low inventory! Risk of stock-out.")
    if row['acos'] > 0.6: # 60% se zyada ACOS
        reasons.append("ACOS is very high, ads are inefficient.")
    if row['organic_rank'] > 50:
        reasons.append("Organic visibility is dropping.")
        
    # Positive Anomalies (Good News)
    if row['sales_change_1d'] > 0.5:
        reasons.append("Sales surge detected! High demand.")
    if row['cvr'] > 0.4:
        reasons.append("Conversion rate is exceptionally high.")

    return " | ".join(reasons) if reasons else "General market fluctuation"

# Anomalies par apply karein
detected_anomalies['alert_reason'] = detected_anomalies.apply(explain_anomaly, axis=1)
```

```python
def explain_anomaly_pro(row):
    reasons = []
    
    # 🚨 CRITICAL: Inventory issues
    if row['inventory_level'] <= 0:
        reasons.append("❌ STOCK OUT: Sales stopped due to zero inventory.")
    elif row['inventory_level'] < 20:
        reasons.append(f"⚠️ LOW STOCK: Only {row['inventory_level']} units left.")

    # 📉 PERFORMANCE: Sales & ACOS
    if row['sales_change_1d'] < -0.25:
        reasons.append(f"📉 SALES DROP: Revenue fell by {abs(row['sales_change_1d']):.1%}.")
    
    if row['acos'] > 0.50:
        reasons.append(f"💸 AD BLEEDING: ACOS is very high ({row['acos']:.1%}). Budget wasting!")

    # 🏆 MARKET: Competition & Rank
    if row['price_gap'] > 50:
        reasons.append(f"💰 PRICE GAP: You are ₹{row['price_gap']:.2f} more expensive than competitors.")
    
    if row['organic_rank'] > 60:
        reasons.append(f"🔍 VISIBILITY LOST: Organic rank slipped to #{row['organic_rank']}.")

    # 🚀 POSITIVE: Growth
    if row['sales_change_1d'] > 0.4:
        reasons.append(f"🔥 SALES SURGE: Growth of {row['sales_change_1d']:.1%}. Check if stock can handle this!")

    return " | ".join(reasons) if reasons else "Neutral Anomaly (Data Variance)"

# Apply upgraded function
detected_anomalies['alert_reason'] = detected_anomalies.apply(explain_anomaly_pro, axis=1)
```
