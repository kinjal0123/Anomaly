```python
import pandas as pd
import numpy as np
import requests

np.random.seed(42)

keywords = ["Aloe vera juice", "Fruit drink", "Curcumin C3", "Graviola", "Ecoensulin"]

rows_per_keyword = 400  
dates = pd.date_range(end=pd.Timestamp.today(), periods=rows_per_keyword)


rows = []

for keyword in keywords:
    base_impressions = np.random.randint(1000, 5000)
    inventory = np.random.randint(50,150)

    for i in range(rows_per_keyword):
        trend = 1 + (i * 0.001)  
        impressions = int(base_impressions * trend * np.random.uniform(0.9,1.1))
        ctr = np.random.uniform(0.02, 0.08)
        clicks = int(impressions * ctr)
        cpc = np.random.uniform(5,15)
        spend = clicks * cpc
        cvr = np.random.uniform(0.1,0.3)
        orders = int(clicks * cvr)
        avg_order_value = np.random.uniform(300,800)
        sales = orders * avg_order_value
        bid = cpc * np.random.uniform(0.9,1.2)
        acos = spend / sales if sales>0 else 0
        roas = sales / spend if spend>0 else 0
        inventory -= orders
        stock_out_flag = 1 if inventory <= 0 else 0
        inventory = max(inventory,0)
        organic_rank = np.random.randint(1,100)
        rows.append([
            dates[i], keyword, impressions, clicks, spend, sales,
            orders, ctr, cpc, cvr, acos, roas, bid,
            inventory, stock_out_flag, organic_rank
        ])

df = pd.DataFrame(rows, columns=[
    "date","keyword","impressions","clicks","spend","sales",
    "orders","ctr","cpc","cvr","acos","roas","bid",
    "inventory_level","stock_out_flag","organic_rank"
])


df = df.sort_values(by=["keyword","date"])
for col in ["impressions","clicks","spend","sales","orders","ctr","cvr","acos","cpc"]:
    df[f"{col}_change_1d"] = df.groupby("keyword")[col].pct_change()


for col in ["ctr","cvr","acos","spend","sales"]:
    df[f"{col}_7d_avg"] = df.groupby("keyword")[col].rolling(7).mean().reset_index(0,drop=True)

df = df.fillna(0)
```

```python
API_KEY = "41CCD23CF84449F6B7F0CA75B4D1DC1B"

def fetch_rainforest_data(keyword):
    url = "https://api.rainforestapi.com/request"
    params = {
        "api_key": API_KEY,
        "type": "search",
        "amazon_domain": "amazon.in",
        "search_term": keyword
    }
    try:
        response = requests.get(url, params=params)
        data = response.json()
        
        
        all_results = data.get("search_results", [])
        
        sponsored_count = 0
        organic_count = 0
        prices = []
        ratings = []

        for item in all_results:
            
            if item.get("sponsored", False):
                sponsored_count += 1
            else:
                organic_count += 1
                
                if "price" in item and "value" in item["price"]:
                    prices.append(item["price"]["value"])
                if "rating" in item:
                    ratings.append(item["rating"])

        avg_price = np.mean(prices) if prices else 0
        avg_rating = np.mean(ratings) if ratings else 0
        
        return {
            "competitor_avg_price": avg_price,
            "sponsored_count": sponsored_count,
            "organic_count": organic_count,
            "competitor_avg_rating": avg_rating
        }
    except Exception as e:
        print(f"Error fetching {keyword}: {e}")
        return {
            "competitor_avg_price": 0, "sponsored_count": 0,
            "organic_count": 0, "competitor_avg_rating": 0
        }
```

```python
market_data = {}
for kw in keywords:
    print(f"Fetching competitor data for {kw}...")
    market_data[kw] = fetch_rainforest_data(kw)
```

```python
market_df = pd.DataFrame.from_dict(market_data, orient='index').reset_index()
market_df.rename(columns={"index":"keyword"}, inplace=True)
```

```python
combined_df = df.merge(market_df, on="keyword", how="left")


combined_df["competition_intensity"] = combined_df["sponsored_count"] / (
    combined_df["sponsored_count"] + combined_df["organic_count"]
)
combined_df["price_gap"] = combined_df["bid"] - combined_df["competitor_avg_price"]
combined_df["market_pressure_score"] = combined_df["cpc_change_1d"].fillna(0) + combined_df["competition_intensity"]

combined_df.fillna(0, inplace=True)


combined_df.to_csv("final_2000_combined_dataset.csv", index=False)
print("Final combined dataset (~2000 rows) ready for anomaly detection!")
```
